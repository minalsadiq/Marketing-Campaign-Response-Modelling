# Business Problem Statement

Marketing campaigns are expensive, and contacting every customer is inefficient.

The objective of this project is to develop a machine learning model that predicts whether a customer will subscribe to a term deposit after a direct marketing campaign.

The model will help financial institutions:

- Improve campaign targeting
- Reduce marketing costs
- Increase conversion rates
- Maximize Return on Investment (ROI)
- Support data-driven decision making
